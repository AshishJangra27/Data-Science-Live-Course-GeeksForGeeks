```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dfx = pd.read_csv('lregre.csv')
x = dfx.values[:,0]
y = dfx.values[:,1]
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
X = (x - x.mean())/x.std()

Y = y

plt.scatter(X,Y,c=Y)
plt.show()

#This step normalises the data 
```

```python
lr = LinearRegression()
lr.fit(X,Y)
```

```python
print([lr.intercept_[0],lr.coef_[0]])
#This is similar to the final thete variable in Previous notebook
```

```python
xtest = np.linspace(-2,2,10)
#This gives us 10 equidistant points between -2 and 2 
print(xtest)
```

```python
def h_theta(x, theta):
    return theta[0] + theta[1]*x
theta = [lr.intercept_[0],lr.coef_[0]]
```

```python
plt.scatter(X,Y)
plt.plot(xtest,h_theta(xtest,theta), color = 'red')
plt.show()
```
