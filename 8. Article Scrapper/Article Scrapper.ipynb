```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.geeksforgeeks.org/taking-input-in-python/')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup
```

```python
soup.find_all('div')
```

## Finding Article Writing Area

```python
s = soup.find('div', class_ = 'article--viewer_content')
```

## Finding Article Headline

```python
s.find('h1').text
```

```python
s.find('div', class_ = 'media')
```

```python
s.find('div', class_ = 'media').text
```

## Article Basic Info

```python
print(s.find('div', class_ = 'media').text)
```

```python
s.find('div', class_ = 'media').find_all('span')
```

```python
for i in s.find('div', class_ = 'media').find_all('span'):
    print(i.text)
```

## Article Content

```python
for i in soup.find_all('div', class_ = 'text'):
    print(i.text)
```

```python
for i in soup.find_all('div', class_ = 'text')[4:]:
    print(i.text)
```

## Finding Code Area

```python
soup.find('td', class_ = 'code')
```

```python
print(soup.find('td', class_ = 'code').text)
```

```python

for i in soup.find_all('td', class_ = 'code'):

    print(i.text)
```

```python
for i in soup.find('td', class_ = 'code').find_all('div'):
    print(i.text)
```

## Printing All Codes on the Article Page

```python
for codes in soup.find_all('td', class_ = 'code'):
    
    for code in codes.find_all('div'):
        print(code.text)
        
    print('-'*50)
```
